### Valeurs aberrantes

```python
from scipy.stats import shapiro

rendements_anormaux = []
alpha = 0.05 #niveau de signification
for col in df.columns[2:]:
    rendement = df[col]
    stat, p = shapiro(rendement)
    if p < alpha:
        rendements_anormaux.append(col)
    else:
        print(f'{col} suit une distribution normale')

print(f'Les rendements qui ne suivent pas une distribution normale sont: {rendements_anormaux}')
```

#### Methode de seuil simple qui s'applique sur les rendements qui suivent une distribution normale ou pas 

```python
# Paramtres pour dfinir les valeurs aberrantes
seuil = 3

# Initialisation de la liste pour stocker les pourcentages de valeurs aberrantes
percentages = []

# Boucle sur les colonnes
for col in df.columns[2:]:
    rendement = df[col]
    moy = rendement.mean()
    std = rendement.std()
    limite_inf = moy - seuil * std
    limite_sup = moy + seuil * std
    valeurs_aberrantes = rendement[(rendement < limite_inf) | (rendement > limite_sup)]
    pourcentage_aberrant = len(valeurs_aberrantes) / len(rendement) * 100
    percentages.append([col, pourcentage_aberrant])

# Trier la liste des pourcentages de valeurs aberrantes
percentages.sort(key=lambda x: x[1], reverse=True)

# Imprimer les variables avec le plus grand nombre de valeurs aberrantes
for var, perc in percentages:
    print(f"La variable {var} a {perc:.2f}% de valeurs aberrantes")
```

```python
# Afficher les plots des 10 variables avec le plus de valeurs aberrantes
for i, (var, perc) in enumerate(percentages[:10]):
    plt.subplot(2, 5, i + 1)
    plt.boxplot(df[var], showfliers=True)
    plt.xlabel('Variable')
    plt.ylabel(var)
    plt.title(f'({perc:.2f}% VA)')
    if i == 9:
        break

plt.tight_layout()
plt.show()
```

```python
# Imputation des valeurs aberrantes par la mediane
import numpy as np

seuil = 3

for col in df.columns[2:]:
    rendement = df[col]
    median = rendement.median()
    moy = rendement.mean()
    std = rendement.std()
    limite_inf = moy - seuil * std
    limite_sup = moy + seuil * std
    rendement[(rendement < limite_inf) | (rendement > limite_sup)] = median
    df[col] = rendement
```

```python
import math

df_sharpe = pd.DataFrame(columns=['Titres', 'Sharpe_ratio'])
rf = df_rf['Price'].mean()

temp = []
for col in df.columns[2:]:
    mean_return = df[col].mean()
    volatilite = df[col].std()
    sharpe_ratio = (mean_return - rf) / volatilite
    temp.append({'Titres': col, 'Sharpe_ratio': sharpe_ratio})

df_sharpe = pd.concat([df_sharpe, pd.DataFrame(temp)], ignore_index=True)
```

### Treynor ratio

```python
import pandas as pd

# Initialiser un dataframe vide pour stocker les ratios de Treynor
df_treynor = pd.DataFrame(columns=['Titres', 'Treynor_ratio'])

# Obtenir le rendement de l'indice de rfrence
rf = df['S5ENRS Index'].mean()

# Boucler sur les colonnes de rendement des titres
for col in df.columns[2:]:
    # Calculer le ratio de Treynor pour chaque titre
    mean_return = df[col].mean()
    beta = df[col].cov(df['S5ENRS Index']) / df['S5ENRS Index'].var()
    treynor_ratio = (mean_return - rf) / beta
    
    # Ajouter le ratio de Treynor et le nom de l'action au dataframe
    temp_df = pd.DataFrame({'Titres': [col], 'Treynor_ratio': [treynor_ratio]})
    df_treynor = pd.concat([df_treynor, temp_df], ignore_index=True)
```

```python
import matplotlib.pyplot as plt

cmap = plt.cm.get_cmap('Dark2')

for i in range(len(df_treynor)):
    plt.bar(i, df_treynor['Treynor_ratio'][i])

plt.legend(df_treynor['Titres'], loc='lower center', bbox_to_anchor=(0.5, -0.9), ncol=5)

plt.title("Ratio de Treynor des titres")
plt.xlabel("Actions")
plt.ylabel("Ratio de Treynor")
plt.show()
```

### VaR historique

```python
#VaR  95% de confiance
confidence = 0.05
var_df = pd.DataFrame(columns=['Titres', 'VaR'])
for col in df.columns[2:]:
    worst_return = df[col].quantile(confidence)
    mean_return = df[col].mean()
    var = mean_return - worst_return
    temp_df = pd.DataFrame({'Titres': [col], 'VaR': [var]})
    var_df = pd.concat([var_df, temp_df])
```

### Omega

```python
# Dfinir le seuil de rendement et le taux sans risque
seuil = 3
taux_sans_risque = df_rf['Price'].mean()

omega_0 = {}
for col in df.columns[2:]:
    rendements_surpassant_seuil = df[col][df[col] > seuil]
    rendements_surpassant_seuil -= taux_sans_risque
    omega_0[col] = rendements_surpassant_seuil.mean() / rendements_surpassant_seuil.std()

omega_0 = pd.DataFrame.from_dict(omega_0, orient='index', columns=["Omega_0"])
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

df_folio=df.copy()

# Slectionner les actifs souhaits
titres = df_strat_filtered['Titres']
df_folio = df_folio[titres.tolist()]

# Calculer les rendements moyens
rendements_moyens = df_folio.mean()

# Calculer la matrice de covariance
matrice_covariance = df_folio.cov()

# Dfinir les contraintes pour la fonction d'optimisation
def contraintes(poids):
    # Somme des poids doit tre gale  1
    return np.sum(poids) - 1

# Dfinir la fonction objectif pour la fonction d'optimisation
def objectif(poids, rendements_moyens, matrice_covariance):
    portfolio_rendement = np.sum(rendements_moyens * poids)
    portfolio_volatilite = np.sqrt(np.dot(poids.T, np.dot(matrice_covariance, poids)))
    return -portfolio_rendement / portfolio_volatilite

# Dfinir les bornes pour les poids
bornes = [(0,1) for i in range(len(titres))]

# Dfinir les poids initiaux pour la fonction d'optimisation
poids_initiaux = [1. / len(titres) for i in range(len(titres))]

# Optimiser les poids pour minimiser la fonction objectif
resultat = minimize(fun=objectif, x0=poids_initiaux, args=(rendements_moyens, matrice_covariance), method="SLSQP", bounds=bornes, constraints={"type": "eq", "fun": contraintes})
poids_optimaux = resultat["x"]

# Diviser le dataframe en donnes d'entranement et de test
split_index = int(df_folio.shape[0] * 0.8)
df_train = df_folio[:split_index]
df_test = df_folio[split_index:]

# Calculer les rendements moyens pour les donnes d'entranement
rendements_moyens_train = df_train.mean()

# Calculer les rendements du portefeuille pour les donnes de test
rendements_portefeuille_test = df_test.dot(poids_optimaux)

# Calculer la performance du portefeuille pour les donnes de test
performance_portefeuille_test = np.mean(rendements_portefeuille_test)

# Calculer la volatilit du portefeuille pour les donnes de test
volatilite_portefeuille_test = np.std(rendements_portefeuille_test)

# Calculer le ratio de Sharpe du portefeuille pour les donnes de test
ratio_sharpe_test = performance_portefeuille_test / volatilite_portefeuille_test

print("Performance du portefeuille sur le jeu de donnes de test:", performance_portefeuille_test)
print("Volatilit du portefeuille sur le jeu de donnes de test:", volatilite_portefeuille_test)
print("Ratio de Sharpe du portefeuille sur le jeu de donnes de test:", ratio_sharpe_test)
```
